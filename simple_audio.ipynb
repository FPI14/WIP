This tutorial demonstrates how to preprocess audio files in the WAV format and build and train a basic [automatic speech recognition](https://en.wikipedia.org/wiki/Speech_recognition) (ASR) model for recognizing ten different words. You will use a portion of the [Speech Commands dataset](https://www.tensorflow.org/datasets/catalog/speech_commands) ([Warden, 2018](https://arxiv.org/abs/1804.03209)), which contains short (one-second or less) audio clips of commands, such as "down", "go", "left", "no", "right", "stop", "up" and "yes".

Real-world speech and audio recognition [systems](https://ai.googleblog.com/search/label/Speech%20Recognition) are complex. But, like [image classification with the MNIST dataset](../quickstart/beginner.ipynb), this tutorial should give you a basic understanding of the techniques involved.

The audio clips are 1 second or less at 16kHz. The `output_sequence_length=16000` pads the short ones to exactly 1 second (and would trim longer ones) so that they can be easily batched.

```
train_ds, val_ds = tf.keras.utils.audio_dataset_from_directory(
    directory=data_dir,
    batch_size=64,
    validation_split=0.2,
    seed=0,
    output_sequence_length=16000,
    subset='both')

label_names = np.array(train_ds.class_names)
print()
print("label names:", label_names)
```

The **batch size** of 64 is a common choice in deep learning for several reasons:

1. **Computational Efficiency**: A batch size of 64 allows for efficient utilization of hardware, especially GPUs. Processing multiple samples at once can lead to faster training times compared to smaller batch sizes, as computations can be parallelized.
2. **Training Stability**: It provides a good balance between the noise introduced by smaller batch sizes and the computational cost and generalization issues of very large batch sizes. Smaller batch sizes (e.g., 1 or 4) can lead to very noisy gradients, which might make training unstable. Larger batch sizes (e.g., 256 or 512) can sometimes lead to models that generalize less effectively to unseen data, and they require more memory.
3. **Memory Management**: A batch size of 64 typically fits comfortably within the memory limits of most GPUs, preventing 'out of memory' errors that can occur with excessively large batches.
4. **Common Practice**: It's a widely adopted and empirically proven batch size in many deep learning research papers and tutorials, making it a good default starting point for many tasks.

(`train_ds.element_spec`):

- **Tenseur audio** (`TensorSpec(shape=(None, 16000, None), dtype=tf.float32, name=None)`):
  
  - Le premier `None` indique la taille du lot, ce qui signifie que le nombre d'échantillons audio dans un lot peut varier.

  - Le deuxième `None` indique le nombre de canaux. Initialement, il est `1` pour l'audio mono, mais après l'opération `tf.squeeze` (dans la cellule `Xl-tnniUIBlM`), cette dimension est supprimée, résultant en une forme `(None, 16000)` pour le tenseur audio.
  - `dtype=tf.float32` signifie que les données audio sont représentées par des nombres à virgule flottante de 32 bits.
- **Tenseur d'étiquettes** (`TensorSpec(shape=(None,), dtype=tf.int32, name=None)`):
  
  - `None` indique la taille du lot, correspondant au nombre d'échantillons audio dans le lot.
  - La forme `()` indique qu'il s'agit d'un tenseur 1D. Chaque élément de ce tenseur est un entier représentant l'étiquette de classe (par exemple, 0 pour 'down', 1 pour 'go', etc.) pour le clip audio correspondant dans le lot.
  - `dtype=tf.int32` signifie que les étiquettes sont représentées par des entiers de 32 bits.

This dataset only contains single channel audio, so use the `tf.squeeze` function to drop the extra axis:

```
def squeeze(audio, labels):
  audio = tf.squeeze(audio, axis=-1)
  return audio, labels

train_ds = train_ds.map(squeeze, tf.data.AUTOTUNE)
val_ds = val_ds.map(squeeze, tf.data.AUTOTUNE)
```

Cette ligne de code applique la fonction `squeeze` à chaque élément du jeu de données `train_ds`. Voici ce que cela signifie :

- **`train_ds.map(squeeze, ...)`** : La méthode `.map()` d'un `tf.data.Dataset` est utilisée pour appliquer une fonction de transformation à chaque élément du jeu de données. Dans ce cas, la fonction `squeeze` est appliquée à chaque paire `(audio, labels)` dans `train_ds`.
  
- **`squeeze` fonction** : La fonction `squeeze` (définie précédemment dans la cellule `Xl-tnniUIBlM`) prend les tenseurs audio et d'étiquettes comme entrée. Elle utilise `tf.squeeze(audio, axis=-1)` pour supprimer la dimension du canal de l'audio. Initialement, l'audio a une forme `(batch_size, 16000, 1)` (où le `1` est la dimension du canal pour l'audio mono). `tf.squeeze` supprime cette dimension de taille 1, transformant la forme en `(batch_size, 16000)`. Ceci est utile car de nombreux modèles (et les étapes suivantes de prétraitement, comme la création de spectrogrammes) n'ont pas besoin de cette dimension de canal explicite pour l'audio mono.
  
- **`tf.data.AUTOTUNE`** : C'est un paramètre qui permet à TensorFlow d'ajuster dynamiquement le nombre de threads de travail à utiliser pour le prétraitement en parallèle. Il optimise les performances en déterminant le meilleur niveau de parallélisme en fonction des ressources disponibles, ce qui peut accélérer le chargement et le prétraitement des données.

## Convert waveforms to spectrograms

The waveforms in the dataset are represented in the time domain. Next, you'll transform the waveforms from the time-domain signals into the time-frequency-domain signals by computing the [short-time Fourier transform (STFT)](https://en.wikipedia.org/wiki/Short-time_Fourier_transform) to convert the waveforms to as [spectrograms](https://en.wikipedia.org/wiki/Spectrogram), which show frequency changes over time and can be represented as 2D images. You will feed the spectrogram images into your neural network to train the model.

A Fourier transform (`tf.signal.fft`) converts a signal to its component frequencies, but loses all time information. In comparison, STFT (`tf.signal.stft`) splits the signal into windows of time and runs a Fourier transform on each window, preserving some time information, and returning a 2D tensor that you can run standard convolutions on.

Create a utility function for converting waveforms to spectrograms:

- The waveforms need to be of the same length, so that when you convert them to spectrograms, the results have similar dimensions. This can be done by simply zero-padding the audio clips that are shorter than one second (using `tf.zeros`).
- When calling `tf.signal.stft`, choose the `frame_length` and `frame_step` parameters such that the generated spectrogram "image" is almost square. For more information on the STFT parameters choice, refer to [this Coursera video](https://www.coursera.org/lecture/audio-signal-processing/stft-2-tjEQe) on audio signal processing and STFT.
- The STFT produces an array of complex numbers representing magnitude and phase. However, in this tutorial you'll only use the magnitude, which you can derive by applying `tf.abs` on the output of `tf.signal.stft`.

```
def get_spectrogram(waveform):
  # Convert the waveform to a spectrogram via a STFT.
  spectrogram = tf.signal.stft(
      waveform, frame_length=255, frame_step=128)
  # Obtain the magnitude of the STFT.
  spectrogram = tf.abs(spectrogram)
  # Add a `channels` dimension, so that the spectrogram can be used
  # as image-like input data with convolution layers (which expect
  # shape (`batch_size`, `height`, `width`, `channels`).
  spectrogram = spectrogram[..., tf.newaxis]
  return spectrogram
```

Le spectrogramme a une forme de **3 dimensions** parce que le code le transforme spécifiquement pour qu'il ressemble à une image en niveaux de gris pour un réseau neuronal convolutif (CNN). Voici la signification de chaque dimension pour une seule forme d'onde:

1. **Première dimension (124)**: Représente le nombre de "fenêtres temporelles" ou de "cadres" sur lesquels la Transformée de Fourier à court terme (STFT) a été calculée. Elle indique comment le signal change au fil du temps.
2. **Deuxième dimension (129)**: Représente le nombre de "bandes de fréquence" ou de "bins de fréquence". Elle indique les différentes fréquences présentes dans le signal.
3. **Troisième dimension (1)**: Cette dimension est ajoutée explicitement par l'opération `spectrogram = spectrogram[..., tf.newaxis]` dans la fonction `get_spectrogram`. Elle agit comme une dimension de "canal", similaire au canal unique (niveaux de gris) d'une image. Les modèles CNN s'attendent généralement à une entrée sous la forme `(hauteur, largeur, canaux)`, donc cette dimension est ajoutée pour répondre à cette exigence.

Now, define a function for displaying a spectrogram:

```
def plot_spectrogram(spectrogram, ax):
  if len(spectrogram.shape) > 2:
    assert len(spectrogram.shape) == 3
    spectrogram = np.squeeze(spectrogram, axis=-1)
  # Convert the frequencies to log scale and transpose, so that the time is
  # represented on the x-axis (columns).
  # Add an epsilon to avoid taking a log of zero.
  log_spec = np.log(spectrogram.T + np.finfo(float).eps)
  height = log_spec.shape[0]
  width = log_spec.shape[1]
  X = np.linspace(0, np.size(spectrogram), num=width, dtype=int)
  Y = range(height)
  ax.pcolormesh(X, Y, log_spec)
```

La fonction `np.squeeze(spectrogram, axis=-1)` est utilisée pour supprimer les dimensions de taille 1 d'un tableau NumPy. Dans ce cas, `axis=-1` indique à `np.squeeze` de ne supprimer que la *dernière* dimension si sa taille est 1.

Now, create spectrogram datasets from the audio datasets:

```
def make_spec_ds(ds):
  return ds.map(
      map_func=lambda audio,label: (get_spectrogram(audio), label),
      num_parallel_calls=tf.data.AUTOTUNE)
```

```
train_spectrogram_ds = make_spec_ds(train_ds)
val_spectrogram_ds = make_spec_ds(val_ds)
test_spectrogram_ds = make_spec_ds(test_ds)
```

## Build and train the model

Add `Dataset.cache` and `Dataset.prefetch` operations to reduce read latency while training the model:

```
train_spectrogram_ds = train_spectrogram_ds.cache().shuffle(10000).prefetch(tf.data.AUTOTUNE)
val_spectrogram_ds = val_spectrogram_ds.cache().prefetch(tf.data.AUTOTUNE)
test_spectrogram_ds = test_spectrogram_ds.cache().prefetch(tf.data.AUTOTUNE)
```

[Keras 3.0 Getting started](https://keras.io/getting_started/)  
[Introduction to Keras for engineers](https://keras.io/getting_started/intro_to_keras_for_engineers/)  


For the model, you'll use a simple convolutional neural network (CNN), since you have transformed the audio files into spectrogram images.

Your `tf.keras.Sequential` model will use the following Keras preprocessing layers:

- `tf.keras.layers.Resizing`: to downsample the input to enable the model to train faster.
- `tf.keras.layers.Normalization`: to normalize each pixel in the image based on its mean and standard deviation.

For the `Normalization` layer, its `adapt` method would first need to be called on the training data in order to compute aggregate statistics (that is, the mean and the standard deviation).

```
input_shape = example_spectrograms.shape[1:]
print('Input shape:', input_shape)
num_labels = len(label_names)

# Instantiate the `tf.keras.layers.Normalization` layer.
norm_layer = layers.Normalization()
# Fit the state of the layer to the spectrograms
# with `Normalization.adapt`.
norm_layer.adapt(data=train_spectrogram_ds.map(map_func=lambda spec, label: spec))

model = models.Sequential([
    layers.Input(shape=input_shape),
    # Downsample the input.
    layers.Resizing(32, 32),
    # Normalize.
    norm_layer,
    layers.Conv2D(32, 3, activation='relu'),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.25),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_labels),
])

model.summary()
```

Configure the Keras model with the Adam optimizer and the cross-entropy loss:

```
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'],
)
```

Train the model over 10 epochs for demonstration purposes:

```
EPOCHS = 10
history = model.fit(
    train_spectrogram_ds,
    validation_data=val_spectrogram_ds,
    epochs=EPOCHS,
    callbacks=tf.keras.callbacks.EarlyStopping(verbose=1, patience=2),
)
```

Let's plot the training and validation loss curves to check how your model has improved during training:

```
metrics = history.history
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(history.epoch, metrics['loss'], metrics['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.ylim([0, max(plt.ylim())])
plt.xlabel('Epoch')
plt.ylabel('Loss [CrossEntropy]')

plt.subplot(1,2,2)
plt.plot(history.epoch, 100*np.array(metrics['accuracy']), 100*np.array(metrics['val_accuracy']))
plt.legend(['accuracy', 'val_accuracy'])
plt.ylim([0, 100])
plt.xlabel('Epoch')
plt.ylabel('Accuracy [%]')
```

## Export the model with preprocessing

The model's not very easy to use if you have to apply those preprocessing steps before passing data to the model for inference. So build an end-to-end version:

```
class ExportModel(tf.Module):
  def __init__(self, model):
    self.model = model

    # Accept either a string-filename or a batch of waveforms.
    # You could add additional signatures for a single wave, or a ragged-batch.
    self.__call__.get_concrete_function(
        x=tf.TensorSpec(shape=(), dtype=tf.string))
    self.__call__.get_concrete_function(
       x=tf.TensorSpec(shape=[None, 16000], dtype=tf.float32))


  @tf.function
  def __call__(self, x):
    # If they pass a string, load the file and decode it.
    if x.dtype == tf.string:
      x = tf.io.read_file(x)
      x, _ = tf.audio.decode_wav(x, desired_channels=1, desired_samples=16000,)
      x = tf.squeeze(x, axis=-1)
      x = x[tf.newaxis, :]

    x = get_spectrogram(x)
    result = self.model(x, training=False)

    class_ids = tf.argmax(result, axis=-1)
    class_names = tf.gather(label_names, class_ids)
    return {'predictions':result,
            'class_ids': class_ids,
            'class_names': class_names}
```

Test run the "export" model:

```
export = ExportModel(model)
export(tf.constant(str(data_dir/'no/01bb6a2a_nohash_0.wav')))
```

Save and reload the model, the reloaded model gives identical output:

```
tf.saved_model.save(export, "saved")
imported = tf.saved_model.load("saved")
imported(waveform[tf.newaxis, :])
```
